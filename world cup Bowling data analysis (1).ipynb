# Data Analysis

```python
player_name = data.loc[data['runs'].idxmax(),'player']
team_name = data.loc[data['runs'].idxmax(),'team']
highest_runs = data.loc[data['runs'].idxmax(),'runs']
opponent = data.loc[data['runs'].idxmax(),'opponent'] 

print("--------> Highest runs were given by the bowler ",player_name," from the team of ",team_name," in one spell which are ",
     highest_runs," against team ",opponent)

player_name = data.loc[data['wickets'].idxmax(),'player']
team_name = data.loc[data['wickets'].idxmax(),'team']
highest_wickets = data.loc[data['wickets'].idxmax(),'wickets']
opponent = data.loc[data['wickets'].idxmax(),'opponent'] 

print("--------> Highest wickets were taken by the bowler ",player_name," from the team of ",team_name," in one spell which are ",
     highest_wickets," against team ",opponent)

player_name = data.loc[data['6s'].idxmax(),'player']
team_name = data.loc[data['6s'].idxmax(),'team']
highest_6s = data.loc[data['6s'].idxmax(),'6s']
opponent = data.loc[data['6s'].idxmax(),'opponent'] 

print("--------> Highest 6s were given by the bowler ",player_name," from the team of ",team_name," in one spell which are ",
     highest_6s," against team ",opponent)

player_name = data.loc[data['4s'].idxmax(),'player']
team_name = data.loc[data['4s'].idxmax(),'team']
highest_4s = data.loc[data['4s'].idxmax(),'4s']
opponent = data.loc[data['4s'].idxmax(),'opponent'] 

print("--------> Highest 4s were given by the bowler ",player_name," from the team of ",team_name," in one spell which are ",
     highest_4s," against team ",opponent)

player_name = data.loc[data['maidens'].idxmax(),'player']
team_name = data.loc[data['maidens'].idxmax(),'team']
highest_maidens = data.loc[data['maidens'].idxmax(),'maidens']
opponent = data.loc[data['maidens'].idxmax(),'opponent'] 

print("--------> Highest madiens were overed by the bowler ",player_name," from the team of ",team_name," in one spell which are ",
     highest_maidens," against team ",opponent)
```

```python
data_players_overs = data.groupby('player').agg({'overs':'sum','maidens':'sum','runs':'sum','wickets':'sum','run_rate':'mean'
,'4s':'sum','6s':'sum','wd':'sum'}).reset_index()
```

```python
sorted_data_players_overs = data_players_overs.sort_values(by='overs', ascending=False).reset_index()
```

```python
sorted_data_players_overs
```

# plot the bar graph between overs and players

```python
data_players_over_top_30 = sorted_data_players_overs.head(30)

plt.figure(figsize=(20,6))
plt.bar(data_players_over_top_30['player'],data_players_over_top_30['overs'],color = 'green')
plt.xlabel('Players')
plt.ylabel('Overs')
plt.title('overs by the top_30 players')
plt.xticks(rotation = 60 , ha = 'right')
plt.tight_layout()
plt.show()
```

# Runs given by top_30 players

```python
data_players_over_top_30 = sorted_data_players_overs.head(30)

#plotting
plt.figure(figsize=(20,6))
plt.bar(data_players_over_top_30['player'],data_players_over_top_30['runs'],color = 'violet')
plt.xlabel('Players')
plt.ylabel('runs')
plt.title('runs given by top_30 bowlers')
plt.xticks(rotation = 45 , ha = 'right') #rotate names for better visibility
plt.tight_layout()

#show the plot
plt.show()
```

# Run_rate from top_30 players

```python
data_players_over_top_30 = sorted_data_players_overs.head(30)

#plotting
plt.figure(figsize=(20,6))
plt.bar(data_players_over_top_30['player'],data_players_over_top_30['run_rate'],color = 'orange')
plt.xlabel('Players')
plt.ylabel('run_rate')
plt.title('run_rate by the top_30 players')
plt.xticks(rotation = 60 , ha = 'right') #rotate player name for better visibility
plt.tight_layout()

#show the plot
plt.show()
```

# Wide_balls from top_30 bowlers

```python
data_players_over_top_30 = sorted_data_players_overs.head(30)

#plotting
plt.figure(figsize=(15,6))
plt.bar(data_players_over_top_30['player'],data_players_over_top_30['wd'],color = 'purple')
plt.xlabel('Players')
plt.ylabel('wd')
plt.title('wide_balls from the top_30 players')
plt.xticks(rotation = 45 , ha = 'right') #rotate player name for better visibility
plt.tight_layout()

#show the plot
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(15,6))

index = np.arange(len(data_players_over_top_30))
bars = ax1.bar(index, data_players_over_top_30['overs'], label='overs', color = 'skyblue')

line = ax1.plot(index, data_players_over_top_30['maidens'], label='maidens', color = 'orange', marker = 'o')

ax1.set_xlabel('Player')
ax1.set_ylabel('Overs',color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

plt.title('maiden overs for top_30 players')

plt.xticks(index, data_players_over_top_30['player'], rotation=45, ha='right')

bars_legend = [bar for bar in bars]
line_legend = [line[0]]
plt.legend(bars_legend + line_legend, ['overs', 'maidens '], loc='upper left')


# Show the plot
plt.tight_layout()
plt.show()
```

# Group by Team

```python
data_team_overs = data.groupby('team').agg({'overs':'sum','maidens':'sum','runs':'sum','wickets':'sum','run_rate':'mean','4s':'sum','6s':'sum','wd':'sum'}).reset_index()
data_team_overs
```

# Maiden overs by each team

```python
# Plotting
plt.figure(figsize=(8, 4))
plt.bar(data_team_overs['team'], data_team_overs['maidens'], color='magenta')
plt.xlabel('Team')
plt.ylabel('maidens')
plt.title('Maiden overs by each Team')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```

# Run_Rate by team bowlers

```python
plt.figure(figsize=(15,6))
plt.plot(data_team_overs['team'],data_team_overs['run_rate'], marker = 'o', linestyle = '--')
plt.grid()
plt.gray()
plt.xlabel('Teams')
plt.ylabel('run_rate')
plt.title('run_rate by the team bowlers')
plt.xticks(rotation = 45, ha = 'right')
plt.tight_layout()

plt.show()
```

# wickets taken by each team

```python
plt.figure(figsize = (8,8))
plt.pie(data_team_overs['wickets'],labels = data_team_overs['team'])
plt.title('wickets taken by each team')
plt.show()
```

# runs given by the team bowlers

```python
plt.figure(figsize=(15,6))
plt.plot(data_team_overs['team'], data_team_overs['runs'], marker='o', linestyle='--')
plt.grid()
plt.gray()
plt.xlabel('Teams')
plt.ylabel('runs')
plt.title('runs given by the Team bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
